```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv')

df.head()
```

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

```python
df.head()
```

```python
df_sydney_processed.info()
```

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

# Linear Regression

```python
x_train, x_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(x_train)
scaled_X_test = scaler.transform(x_test)
LinearReg = LinearRegression()
LinearReg.fit(scaled_X_train, y_train)
```

```python
predictions = LinearReg.predict(scaled_X_test)
```

```python
LinearRegression_MAE = metrics.mean_absolute_error(y_test, predictions)
LinearRegression_MSE = metrics.mean_squared_error(y_test, predictions)
LinearRegression_R2 = metrics.r2_score(y_test, predictions)
```

```python
Report = pd.DataFrame({'error_type':['LinearRegression_MAE','LinearRegression_MSE','LinearRegression_R2'],
              'value':[LinearRegression_MAE, LinearRegression_MSE, LinearRegression_R2]})
Report
```

# KNN

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(x_train)
scaled_X_test = scaler.transform(x_test)
KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(scaled_X_train, y_train)
```

```python
predictions = KNN.predict(scaled_X_test)
```

```python
KNN_Accuracy_Score = metrics.accuracy_score(y_test, predictions)
KNN_JaccardIndex = metrics.jaccard_score(y_test, predictions)
KNN_F1_Score = metrics.f1_score(y_test, predictions)
```

```python
KNN_Accuracy_Score
```

```python
KNN_JaccardIndex
```

```python
KNN_F1_Score
```

# DECISION TREE

```python
Tree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
Tree.fit(x_train, y_train)
```

```python
predictions = Tree.predict(x_test)
```

```python
Tree_Accuracy_Score = metrics.accuracy_score(y_test, predictions)
Tree_JaccardIndex = metrics.jaccard_score(y_test, predictions)
Tree_F1_Score = metrics.f1_score(y_test, predictions)
print(Tree_Accuracy_Score)
print(Tree_JaccardIndex)
print(Tree_F1_Score)
```

# Logistic Regression


```python
x_train, x_test, y_train, y_test = train_test_split(features, Y, test_size=0.2, random_state=1)
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(x_train)
scaled_X_test = scaler.transform(x_test)
LR = LogisticRegression().fit(scaled_X_train, y_train)
```

```python
predictions = LR.predict(scaled_X_test)
```

```python
LR_Accuracy_Score = metrics.accuracy_score(y_test, predictions)
LR_JaccardIndex = metrics.jaccard_score(y_test, predictions)
LR_F1_Score = metrics.f1_score(y_test, predictions)
LR_Log_Loss = metrics.log_loss(y_test, predictions)
print(LR_Accuracy_Score)
print(LR_JaccardIndex)
print(LR_F1_Score)
print(LR_Log_Loss)
```

# SVM

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(x_train)
scaled_X_test = scaler.transform(x_test)
SVM = svm.SVC(kernel='linear', C=0.05)
SVM.fit(scaled_X_train, y_train)
```

```python
predictions = SVM.predict(scaled_X_test)
```

```python
SVM_Accuracy_Score = metrics.accuracy_score(y_test, predictions)
SVM_JaccardIndex = metrics.jaccard_score(y_test, predictions)
SVM_F1_Score = metrics.f1_score(y_test, predictions)
print(SVM_Accuracy_Score)
print(SVM_JaccardIndex)
print(SVM_F1_Score)
```

```python
import pandas as pd
from sklearn.metrics import log_loss

y_prob_lr = LR.predict_proba(x_test)
lr_log_loss = log_loss(y_test, y_prob_lr)

Report = pd.DataFrame({
    'Model': ['Linear Regression', 'KNN', 'Decision Tree', 'Logistic Regression', 'SVM'],
    'Accuracy': [LinearRegression_MAE, KNN_Accuracy_Score, Tree_Accuracy_Score, LR_Accuracy_Score, SVM_Accuracy_Score],
    'Jaccard Index': [None, KNN_JaccardIndex, Tree_JaccardIndex, LR_JaccardIndex, SVM_JaccardIndex],
    'F1-Score': [None, KNN_F1_Score, Tree_F1_Score, LR_F1_Score, SVM_F1_Score],
    'LogLoss': [None, None, None, lr_log_loss, None]
})
Report = Report[['Model', 'Accuracy', 'Jaccard Index', 'F1-Score', 'LogLoss']]
Report.style\
    .background_gradient(cmap='Reds', subset=['Accuracy', 'F1-Score'])\
    .format({
        'Accuracy': '{:.2%}',
        'Jaccard Index': '{:.3f}',
        'F1-Score': '{:.3f}',
        'LogLoss': '{:.3f}'
    })
```
